```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Estilo dos gráficos
sns.set(style="whitegrid")
```

```python
df = pd.read_parquet("yellow_tripdata_2025-01.parquet")
```

```python
print("Formato:", df.shape)
print("\nColunas nulas:\n", df.isna().sum())
```

```python
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Criar colunas de data/hora
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_day'] = df['tpep_pickup_datetime'].dt.day_name()
```

```python
df = df[(df['trip_distance'] > 0) & (df['total_amount'] > 0) & (df['tip_amount'] >= 0)]
```

```python
# Códigos: 1=Crédito, 2=Dinheiro, 3=Sem carga, 4=Disputa, 5=Desconhecido, 6=Voided trip
payment_mapping = {
    1: 'Cartão de Crédito',
    2: 'Dinheiro',
    3: 'Sem Cobrança',
    4: 'Disputa',
    5: 'Desconhecido',
    6: 'Corrida Cancelada'
}

df['payment_type_desc'] = df['payment_type'].map(payment_mapping)
```

```python
tip_by_payment = df.groupby('payment_type_desc')['tip_amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=tip_by_payment.index, y=tip_by_payment.values, palette='viridis')
plt.title('💳 Gorjeta Média por Tipo de Pagamento', fontsize=16)
plt.ylabel('Gorjeta Média ($)')
plt.xlabel('Tipo de Pagamento')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Ordem correta dos dias (segunda a domingo)
dias_ordem = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['pickup_day'] = pd.Categorical(df['pickup_day'], categories=dias_ordem, ordered=True)

# Agrupar por dia e hora
corridas_por_hora_dia = df.groupby(['pickup_day', 'pickup_hour']).size().reset_index(name='total_corridas')
```

```python
# Transformar em pivot para heatmap
tabela_heatmap = corridas_por_hora_dia.pivot(index='pickup_day', columns='pickup_hour', values='total_corridas')


plt.figure(figsize=(14, 6))
sns.heatmap(tabela_heatmap, cmap='YlGnBu', linewidths=.5, linecolor='gray', cbar_kws={'label': 'Número de Corridas'})
plt.title('🚖 Volume de Corridas por Hora e Dia da Semana', fontsize=16)
plt.xlabel('Hora do Dia')
plt.ylabel('Dia da Semana')
plt.tight_layout()
plt.show()
```

```python
df['trip_duration_min'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filtrar possíveis outliers
outliers = df[
    (df['trip_distance'] > 50) | 
    (df['trip_distance'] == 0) | 
    (df['fare_amount'] > 200) | 
    (df['fare_amount'] <= 0) |
    (df['trip_duration_min'] <= 0) |
    (df['trip_duration_min'] > 180)
]

print(f"Número de corridas suspeitas encontradas: {len(outliers)}")
outliers[['trip_distance', 'fare_amount', 'trip_duration_min']].describe()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(x=df['trip_distance'])
plt.title('📦 Distribuição da Distância das Corridas (com outliers)')
plt.xlabel('Distância (milhas)')
plt.show()
```

```python
df['is_outlier'] = (
    (df['trip_distance'] > 50) | 
    (df['trip_distance'] == 0) | 
    (df['fare_amount'] > 200) | 
    (df['fare_amount'] <= 0) |
    (df['trip_duration_min'] <= 0) |
    (df['trip_duration_min'] > 180)
)
```

```python
# Agrupar por Localização de Pickup e Dropoff
df_localizacao = df.groupby(['PULocationID', 'DOLocationID']).size().reset_index(name='total_corridas')

# Top 10 Localizações com mais corridas
top_localizacoes = df_localizacao.sort_values(by='total_corridas', ascending=False).head(10)
print(top_localizacoes)

# Plotar as localizações mais populares
plt.figure(figsize=(12, 6))
sns.barplot(x='total_corridas', y='PULocationID', data=top_localizacoes, color='purple')
plt.title('Top 10 Localizações de Pickup mais populares')
plt.xlabel('Número de Corridas')
plt.ylabel('Localização de Pickup')
plt.show()
```
